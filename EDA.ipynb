##### EDA:


```python
def histplot_graph(data,columns):

    for i in columns:
        plt.figure(figsize=(8,4))
        sns.histplot(data[i],kde=True)
        plt.title(f"Distribution of {i}")
        col_mean=data[i].mean()
        col_median=data[i].median()
        plt.axvline(col_mean,color="red",linestyle="--",label="Mean")
        plt.axvline(col_median,color="green",linestyle="--",label="Median")
    
```

```python
histplot_graph(data,columns=["SeniorCitizen",'tenure','MonthlyCharges','TotalCharges'])
```

```python
def histplot_graph(data, columns):

    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    axes = axes.flatten()

    for ax, col in zip(axes, columns):

        sns.histplot(data[col], kde=True, ax=ax)

        mean = data[col].mean()
        median = data[col].median()

        ax.axvline(mean, color="red", linestyle="--")
        ax.axvline(median, color="green", linestyle="--")

        ax.set_title(col)

    plt.tight_layout()
    plt.show()
```

```python
histplot_graph(data,columns=["SeniorCitizen",'tenure','MonthlyCharges','TotalCharges'])
```

```python
def boxplot(data,columns):
    fig,axes=plt.subplots(2,2,figsize=(10,8))
    axes = axes.flatten()
    for ax,i in zip(axes,columns):

        sns.boxplot(y=data[i],ax=ax)
        ax.set_title(f"Box plot of {i}")
        ax.set_ylabel(i)
    plt.show()
```

```python
boxplot(data,columns=["SeniorCitizen",'tenure','MonthlyCharges','TotalCharges'])
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data[["tenure","MonthlyCharges","TotalCharges"]].corr(),annot=True,cmap="coolwarm",fmt=".2f")
```

```python
def count_plot(data,columns):
    fig,axes=plt.subplots(4,2,figsize=(12,14))
    axes=axes.flatten()

    for ax,i in zip(axes,columns):
        sns.countplot(data,x=data[i],ax=ax)
        ax.set_title(f"Countplot of {i}")
        ax.set_ylabel(i)
    plt.show()
```

```python
count_plot(data,columns=data.select_dtypes(include="object"))
```
